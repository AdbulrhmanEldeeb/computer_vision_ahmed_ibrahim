## reinhard color transfer 

```python
import cv2
import numpy as np

def reinhard_color_transfer(source, target):
    """
    Apply Reinhard color transfer from the source to the target image.

    Parameters:
    - source: The source image (reference image).
    - target: The target image (image to be color corrected).

    Returns:
    - output: The target image with colors transferred from the source image.
    """

    def image_stats(image):
        """
        Compute the mean and standard deviation of each channel in the image.
        """
        mean, std = cv2.meanStdDev(image)
        return mean.flatten(), std.flatten()

    # Convert images from BGR to LAB color space
    source_lab = cv2.cvtColor(source, cv2.COLOR_BGR2LAB).astype("float32")
    target_lab = cv2.cvtColor(target, cv2.COLOR_BGR2LAB).astype("float32")

    # Compute the mean and standard deviation of each channel in both images
    source_mean, source_std = image_stats(source_lab)
    target_mean, target_std = image_stats(target_lab)

    # Subtract the mean from the target image
    (l, a, b) = cv2.split(target_lab)
    l -= target_mean[0]
    a -= target_mean[1]
    b -= target_mean[2]

    # Scale by the standard deviation
    l = (target_std[0] / source_std[0]) * l
    a = (target_std[1] / source_std[1]) * a
    b = (target_std[2] / source_std[2]) * b

    # Add the source mean
    l += source_mean[0]
    a += source_mean[1]
    b += source_mean[2]

    # Clip the values to the valid range for LAB color space
    l = np.clip(l, 0, 255)
    a = np.clip(a, 0, 255)
    b = np.clip(b, 0, 255)

    # Merge the channels and convert back to BGR color space
    output_lab = cv2.merge([l, a, b])
    output = cv2.cvtColor(output_lab.astype("uint8"), cv2.COLOR_LAB2BGR)

    return output

# Example usage:
source_image = cv2.imread('images\\sun.jfif')
target_image = cv2.imread('images\\apple.jfif')
result_image = reinhard_color_transfer(source_image, target_image)

# Save or display the result
cv2.imwrite('result.jpg', result_image)
cv2.imshow('Result', result_image)
cv2.imshow('original', target_image)
cv2.imshow('source', source_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## image blending 

```python
import cv2
import numpy as np

# Load the images
apple = cv2.imread('images\\messi.jfif')
orange = cv2.imread('images\\ronaldo.jfif')

# Print the shapes of the images
print(apple.shape)
print(orange.shape)

# Crop the images to the same size
min_height = min(apple.shape[0], orange.shape[0])
min_width = min(apple.shape[1], orange.shape[1])

apple = apple[:min_height, :min_width]
orange = orange[:min_height, :min_width]

# Combine the images horizontally
apple_orange = np.hstack((apple[:, :min_width//2], orange[:, min_width//2:]))

# Generate Gaussian pyramid for apple
apple_copy = apple.copy()
gp_apple = [apple_copy]
for i in range(6):
    apple_copy = cv2.pyrDown(apple_copy)
    gp_apple.append(apple_copy)

# Generate Gaussian pyramid for orange
orange_copy = orange.copy()
gp_orange = [orange_copy]
for i in range(6):
    orange_copy = cv2.pyrDown(orange_copy)
    gp_orange.append(orange_copy)

# Generate Laplacian Pyramid for apple
lp_apple = [gp_apple[5]]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_apple[i])
    gaussian_expanded = cv2.resize(gaussian_expanded, (gp_apple[i-1].shape[1], gp_apple[i-1].shape[0]))
    laplacian = cv2.subtract(gp_apple[i-1], gaussian_expanded)
    lp_apple.append(laplacian)

# Generate Laplacian Pyramid for orange
lp_orange = [gp_orange[5]]
for i in range(5, 0, -1):
    gaussian_expanded = cv2.pyrUp(gp_orange[i])
    gaussian_expanded = cv2.resize(gaussian_expanded, (gp_orange[i-1].shape[1], gp_orange[i-1].shape[0]))
    laplacian = cv2.subtract(gp_orange[i-1], gaussian_expanded)
    lp_orange.append(laplacian)

# Now add left and right halves of images in each level
apple_orange_pyramid = []
for apple_lap, orange_lap in zip(lp_apple, lp_orange):
    cols, rows, ch = apple_lap.shape
    laplacian = np.hstack((apple_lap[:, :cols//2], orange_lap[:, cols//2:]))
    apple_orange_pyramid.append(laplacian)

# Now reconstruct
apple_orange_reconstruct = apple_orange_pyramid[0]
for i in range(1, 6):
    apple_orange_reconstruct = cv2.pyrUp(apple_orange_reconstruct)
    apple_orange_reconstruct = cv2.resize(apple_orange_reconstruct, (apple_orange_pyramid[i].shape[1], apple_orange_pyramid[i].shape[0]))
    apple_orange_reconstruct = cv2.add(apple_orange_pyramid[i], apple_orange_reconstruct)

# Display the images
cv2.imshow("apple", apple)
cv2.imshow("orange", orange)
cv2.imshow("apple_orange", apple_orange)
cv2.imshow("apple_orange_reconstruct", apple_orange_reconstruct)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## vehicle detection and counting 

```python
import cv2 as cv
import time

cap = cv.VideoCapture('videos\\cars_count.mp4')
offset = 5  # Define the margin around the line for counting
y_lim = 550  # Define the location of the horizontal line
sub = cv.createBackgroundSubtractorMOG2(history=100, varThreshold=50)
num_cars = 0

def find_center(x, y, w, h):
    center = (int(x + 0.5 * w), int(y + 0.5 * h))
    return center

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Frame processing
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(gray, (5, 5), 5)
    mask = sub.apply(blur)
    dilated = cv.dilate(mask, (5, 5))
    dilated = cv.morphologyEx(dilated, cv.MORPH_CLOSE, (5, 5))
    
    contours, _ = cv.findContours(dilated, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    
    cv.line(frame, (25, y_lim), (1200, y_lim), (255, 255, 0), 3)  # Draw the horizontal line
    
    current_detections = []
    for c in contours:
        x, y, w, h = cv.boundingRect(c)
        if w < 80 or h < 80:
            continue
        center = find_center(x, y, w, h)
        current_detections.append(center)

        cv.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        cv.circle(frame, center, 3, (0, 0, 255), -1)
    
    # Check if cars are crossing the line
    for c in current_detections:
        if c[1] > y_lim - offset and c[1] < y_lim + offset:
            num_cars += 1
            current_detections.remove(c)
            cv.line(frame, (25, y_lim), (1200, y_lim), (0, 255, 255), 3)  # Change the line color when a car crosses
    
    cv.putText(frame, f'{num_cars} vehicles  passed', (100, 80), cv.FONT_HERSHEY_COMPLEX, 1, (255, 0, 255), 2)

    # Display the frame
    cv.imshow('frame', frame)
    cv.imshow('dilated', dilated)
    
    if cv.waitKey(1) == 27:  # Exit if 'Esc' is pressed
        break

cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv 
shapes=cv.imread('images/shappes.jpg')
shapes_gray=cv.cvtColor(shapes,cv.COLOR_BGR2GRAY)
ret,thresh=cv.threshold(shapes_gray,220,255,cv.THRESH_BINARY)
# Find contours in the thresholded image
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
# cv.drawContours(shapes, contours, -1, (0, 255, 0), 2)

for contour in contours : 
    # cv.arcLength(contour, True) بتحسب طول ال contour و true لو كان كله مقفول 
    approx=cv.approxPolyDP(contour, 0.01* cv.arcLength(contour, True), True)
    x=approx.ravel()[0]-20
    y=approx.ravel()[1]-5 
    cv.drawContours(shapes, [approx], 0, (0, 0, 0), 5)
    if len(approx)==3 : 
        cv.putText(shapes,'trianlge',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)
    elif len(approx)==4 : 
        x,y,w,h=cv.boundingRect(approx)
        aspect_ratio=float(w)/h 
        if aspect_ratio>.95 and aspect_ratio<1.05 : 
            cv.putText(shapes,'square ',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)
        else :
            cv.putText(shapes,'rect ',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)
    elif len(approx)==5  : 
        cv.putText(shapes,'five',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)
    elif len(approx)==10  : 
        cv.putText(shapes,'star',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)
    else : 
        cv.putText(shapes,'circle',(x,y),cv.FONT_HERSHEY_COMPLEX,1,(255,0,0),1)

cv.imshow('shapes',shapes)

cv.waitKey(0)
cv.destroyAllWindows() 
```

## face detection using har cascade 

```python
import numpy as np 
import cv2 
import os 
cap =cv2.VideoCapture(0)

face_cascade =cv2.CascadeClassifier("files\\haarcascade_frontalface_default.xml")
eyes_cascade =cv2.CascadeClassifier("files\\haarcascade_eye.xml")

while(cap.isOpened()):
    ret ,frame =cap.read()
    if not ret : 
        break 
    gray =cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    faces =face_cascade.detectMultiScale(gray ,1.1, 4)
    for (x,y,h,w) in faces :
        cv2.rectangle(frame ,(x,y),(x+w ,y+h),(0,255,0),3)       
        roi_gray=gray[y:y+h ,x:x+w]       
        roi_color=frame[y:y+h ,x:x+w]       
        eyes =eyes_cascade.detectMultiScale(roi_gray)       
        for(ex,ey,eh,ew) in eyes :      
            cv2.rectangle(roi_color ,(ex,ey),(ex+ew ,ey+eh),(0,0,255),3)            
  
    
    cv2.imshow("frame" ,frame)
    
    key =cv2.waitKey(20)
    if(key == 27):
        break 
        
cap.release()
cv2.destroyAllWindows()
```

## dlib 

```python
import dlib
import cv2 as cv

# Initialize face detector and shape predictor
detector = dlib.get_frontal_face_detector()
# download it if it is not downloaded from here https://www.kaggle.com/datasets/sergiovirahonda/shape-predictor-68-face-landmarksdat
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

# Start video capture from the default camera
cap = cv.VideoCapture(0)

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

while cap.isOpened():
    # Read a frame from the camera
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    # Convert the frame to grayscale for face detection
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    # Detect faces in the grayscale image
    faces = detector(gray)
    
    # Iterate over detected faces
    for face in faces:
        # Get the coordinates of the bounding box
        x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()
        num_faces=len(faces)
        # Draw the bounding box around the face
        cv.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 3)

        # Predict the facial landmarks
        landmarks = predictor(gray, face)

        # Draw landmarks on the face
        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            cv.circle(frame, (x, y), 2, (255, 0, 0), -1)  # Draw each landmark as a small circle
            cv.putText(frame,f'{num_faces} faces are in the frame',(25,25),cv.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
    # Display the frame with the detections
    cv.imshow('Face Detection and Landmark Prediction', frame)

    # Check for 'q' key press to exit
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all OpenCV windows
cap.release()
cv.destroyAllWindows()
```

## drowsy driver 

```python
import dlib
import cv2
import time
from scipy.spatial import distance
from imutils import face_utils
import imutils
import pygame  # Alternative to playsound for stability

# Constants
model_path = "shape_predictor_68_face_landmarks.dat"
ALARM_PATH = "videos/assets_alarm.mp3"
thresh = 0.25
frame_check = 10

# Initialize audio
pygame.mixer.init()
def play_alarm():
    pygame.mixer.music.load(ALARM_PATH)
    pygame.mixer.music.play()

def eye_aspect_ratio(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    return (A + B) / (2.0 * C)

def detect():
    # Initialize dlib detector and predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor(model_path)

    # Eye landmarks
    (lStart, lEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["left_eye"]
    (rStart, rEnd) = face_utils.FACIAL_LANDMARKS_68_IDXS["right_eye"]

    # Start video capture
    cap = cv2.VideoCapture(0)
    flag = 0
    ALARM_ON = False

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        frame = imutils.resize(frame, width=640, height=480)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        subjects = detector(gray)

        for subject in subjects:
            shape = predictor(gray, subject)
            shape = face_utils.shape_to_np(shape)

            leftEye = shape[lStart:lEnd]
            rightEye = shape[rStart:rEnd]
            leftEAR = eye_aspect_ratio(leftEye)
            rightEAR = eye_aspect_ratio(rightEye)
            ear = (leftEAR + rightEAR) / 2.0

            cv2.drawContours(frame, [cv2.convexHull(leftEye)], -1, (0, 255, 0), 1)
            cv2.drawContours(frame, [cv2.convexHull(rightEye)], -1, (0, 255, 0), 1)

            if ear < thresh:
                flag += 1
                if flag >= frame_check and not ALARM_ON:
                    ALARM_ON = True
                    play_alarm()
                cv2.putText(frame, "Drowsy!", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
            else:
                flag = 0
                ALARM_ON = False
                pygame.mixer.music.stop()

        cv2.imshow("Drowsiness Detection", frame)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break

    cap.release()
    cv2.destroyAllWindows()
    pygame.mixer.quit()

if __name__ == '__main__':
    detect()
```
